```python
# Define a function to map age to age groups
def age_to_group(age):
    if 25 <= age <= 35:
        return '25 to 35'
    elif 35 < age <= 45:
        return '35 to 45'
    elif 45 < age <= 60:
        return '45 to 60'
    else:
        return 'Other'

# Apply the function to create a new column for age groups
data['Age_Group'] = data['Parent_age'].apply(age_to_group)

# Display the first few rows of the DataFrame with the new column
print(data[['Parent_age', 'Age_Group']].head())
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Age_Group', hue='Comfort_Expressing_Emotions', order=['25 to 35', '35 to 45', '45 to 60'])
plt.title('Comfort Expressing Emotions Among Different Age Groups of Parents')
plt.xlabel('Age Group of Parents')
plt.ylabel('Count')
plt.legend(title='Comfort Level')
plt.show()
```

```python
# Create a cross-tabulation (contingency table) of the variables
cross_tab = pd.crosstab(index=data['Employment_Status'], columns=[data['Bonding_activities_with_child']])

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, cmap='YlGnBu', annot=True, fmt='d', cbar=True)
plt.title('Heatmap of Categorical Variables')
plt.xlabel('Both Parents Working & Bonding Activities')
plt.ylabel('Employment Status')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the data
# Create a pivot table to calculate the average bonding activities with children
pivot_table = data.pivot_table(index='Bonding_activities_with_child', columns='Parenting_Approach', values='Employment_Status_Numeric', aggfunc='mean')

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".1f", linewidths=.5)
plt.title('Average Bonding Activities with Children by Employment Status and Parenting Approach')
plt.xlabel('Parenting Approach')
plt.ylabel('Employment Status')
plt.show()
```
